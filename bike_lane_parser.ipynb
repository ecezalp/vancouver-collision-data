```python
from pyproj import Proj, transform
import pandas as pd
```

```python
df = pd.read_excel('bike_lanes.xlsx')
```

```python
df.head()
```

```python



def convertCoords(row):
    p = Proj(proj='utm',zone=10,ellps='WGS84')
    lon, lat = p(row['POINT_X'],row['POINT_Y'], inverse=True)
    return  pd.Series([lon, lat])
```

```python
df[['newLong','newLat']] = df.apply(convertCoords,axis=1)
```

```python
df.head()
```

```python
df.to_csv('bike_lanes_latlon.csv')
```
